```python
# Load once at the top
from spacy.lang.en import English
import string

nlp = English()

# Clean text function
def clean_text(text):
    # Lowercase
    cleaned = text.lower()

    # Remove punctuation
    cleaned = "".join([char for char in cleaned if char not in string.punctuation])

    # Tokenize
    my_doc = nlp(cleaned)
    token_list = [token.text for token in my_doc]

    # Remove stopwords
    filtered_sentence = []
    for word in token_list:
        lexeme = nlp.vocab[word]
        if not lexeme.is_stop:
            filtered_sentence.append(word)

    # Return as string
    return " ".join(filtered_sentence)
```

```python
data["cleaned"] = data["text"].apply(clean_text)
```

```python
data.head()
```

```python
# Number of words in original text
data["word_count"] = data["text"].apply(lambda x: len(x.split()))

# Number of words in cleaned text
data["word_count_cleaned"] = data["cleaned"].apply(lambda x: len(x.split()))

# Number of characters including spaces in the cleaned text
data["char_count"] = data["cleaned"].apply(lambda x: len(x))

# Number of characters excluding spaces in the cleaned text
data["char_count_without_spaces"] = data["cleaned"].apply(lambda x: len(x.replace(" ", "")))

# Number of digits in the cleaned text
data["num_dig"] = data["cleaned"].apply(lambda x: sum(1 for w in x.split() if w.isdigit()))
```

```python
# Import spaCy English language model
import spacy
nlp = spacy.load("en_core_web_sm")

# Create spaCy document from the first cleaned text entry
document = nlp(data['cleaned'][0])

# View sample document
document

# Extract POS (Part-of-Speech) tags from the document
all_tags = []
for w in document:
    all_tags.append(w.tag_)

# Display the POS tags
all_tags

# Dictionary of POS tags for nouns and verbs
pos_dic = {
    "noun": ["NNP", "NN", "NNS", "NNPS"],
    "verb": ["VBZ", "VB", "VBD", "VBG", "VBN"]
}
```

```python
all_tags
```

```python
# POS dictionary to classify tags
pos_dic = {
    "noun": ["NNP", "NN", "NNS", "NNPS"],
    "verb": ["VBZ", "VB", "VBD", "VBG", "VBN"]
}

# Function to count noun or verb POS tags in a document
def pos_check(txt, family):
    # Create spaCy document
    txt = nlp(txt)

    # Extract all POS tags
    all_tags = []
    for w in txt:
        all_tags.append(w.tag_)

    # Count the number of specified POS tags
    count = 0
    for tag in all_tags:
        if tag in pos_dic[family]:
            count += 1

    return count
```

```python
# Sample document: Verb count
pos_check("They are playing in the ground", "verb")
```

```python
# Sample document: Noun count
pos_check("They are playing in the ground", "noun")
```

```python
# Applying the function to the entire DataFrame
data["noun_count"] = data["cleaned"].apply(lambda x: pos_check(x, "noun"))
data["verb_count"] = data["cleaned"].apply(lambda x: pos_check(x, "verb"))
```

```python
from sklearn.preprocessing import LabelEncoder

target = data["label"].values
target = LabelEncoder().fit_transform(target)
```

```python
train = data[[
    'word_count',
    'word_count_cleaned',
    'char_count',
    'char_count_without_spaces',
    'num_dig',
    'noun_count',
    'verb_count'
]]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split(
    train, target, random_state=20, stratify=target
)
```

```python
from sklearn import naive_bayes

# Initialize the Multinomial Naive Bayes model
model = naive_bayes.MultinomialNB()
```

```python
model.fit(x_train, y_train)
```

```python
# Predictions on training data
pred_train = model.predict(x_train)

# Predictions on validation data
pred_valid = model.predict(x_valid)
```
